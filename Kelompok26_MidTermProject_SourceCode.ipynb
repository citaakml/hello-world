#Import Library


```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

#Dataset dan Pembentukan Graf

*disarankan untuk draw graf dilakukan terakhir karena prosesnya cukup lama

```python
#read dataset
data = pd.read_csv('kuliner-bakso-sate_dataset.csv')
data
```

```python
#create graph
G = nx.DiGraph()
G = nx.from_pandas_edgelist(data, source= 'source', target = 'target', create_using=nx.Graph())
```

```python
pos = nx.spring_layout(G, k = 0.9)
nx.draw_networkx_nodes(G, pos, node_size = 20, node_color = 'green')
nx.draw_networkx_edges(G, pos, alpha = 0.3, edge_color='black')
plt.axis('off')
plt.show()
```

```python
G['FOOD_FESS']
```

#Menghitung Sentralitas

```python
# Calculate degree centrality
degree_centrality = nx.degree_centrality(G)
degree_centrality = sorted(degree_centrality.items(), key=lambda x:x[1], reverse=True)
#print("Degree centrality:", degree_centrality)
degree_cent_df = pd.DataFrame(degree_centrality[:10], columns=['Username','Degree_centrality'])
degree_cent_df
```

```python
print(degree_cent_df.to_latex())
```

```python
# Calculate betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)
betweenness_centrality = sorted(betweenness_centrality.items(), key=lambda x:x[1], reverse=True)
between_cent_df = pd.DataFrame(betweenness_centrality[:10], columns=['Username','Betweenness_centrality'])
between_cent_df
```

```python
print(between_cent_df.to_latex())
```

```python
# Calculate eigenvector centrality
eigenvector_centrality = nx.eigenvector_centrality_numpy(G)
eigenvector_centrality = sorted(eigenvector_centrality.items(), key=lambda x:x[1], reverse=True)
#print("Eigenvector centrality:", eigenvector_centrality)
eigen_cent_df = pd.DataFrame(eigenvector_centrality[:10], columns=['Username','Eigenvector_centrality'])
eigen_cent_df
```

```python
print(eigen_cent_df.to_latex())
```
